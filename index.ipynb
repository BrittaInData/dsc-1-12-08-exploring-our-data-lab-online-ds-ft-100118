### Build normalized histograms with kde plots to explore the distributions further. 
### Use only the continuous variables in the dataset to plot these visualizations. 

```python
for column in ['Weekly_Sales', 'Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']:
    df[column].plot.hist(density=True, bins=20)
    df[column].plot.kde(label=column)
    plt.legend()
    plt.show();
```

```python
# State your observations here 
```

* Kernel density estimation or KDE is a non-parametric way to estimate the probability density function of a random variable. 
* In other words, **the aim of KDE is to find probability density function (PDF) for a given dataset**, how it differs from normalized histogram approach. Well, it smooths the around values of PDF.
* In our dataset: Many variables are multimodal.

### Build joint plots to check for the linearity assumption between predictors and target variable

Let's use a slightly more advanced plotting technique in seaborn that uses scatter plots, distributions, kde and simple regression line - all in a single go. Its called a `jointplot`. [Here is the official doc. for this method](https://seaborn.pydata.org/generated/seaborn.jointplot.html). 

Here is how you would use it:

> **`sns.jointplot(x= <column>, y= <column>, data=<dataset>, kind='reg')`**

A joint plot will allow us to visually inspect linearity as well as normality assumptions as a single step. 

```python
for column in ['Store', 'Dept', 'Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']:
    sns.jointplot(x=column, y='Weekly_Sales', data=df, kind='reg', label=column)
    plt.legend()
    plt.show();
```

```python
# Provide your observations here 
```

* Distributions are known from previous plots: most of them are not normally distributed.
* Multimodality in some features ist also known, `Store` is (unsurprisingly) almost uniform.
* Regarding the regression line: most variables show no clear relationship with sales. Only `Size` with a slightly positive relationship and `CPI` with a slightly negative one.
